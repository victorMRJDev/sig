```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gaussian_kde

ingei_municip = ingei_municip.dropna(subset=['TotalMuni'])
ingei_municip = ingei_municip[ingei_municip['TotalMuni'] != 'ND']

# Convertir la columna 'TotalMuni' a numérica
ingei_municip['TotalMuni'] = pd.to_numeric(ingei_municip['TotalMuni'], errors='coerce')

# Filtrar los municipios para excluir aquellos con valores no numéricos en la columna 'TotalMuni'
top_municipios = ingei_municip.head(10)

top_municipios = ingei_municip.nlargest(5, 'TotalMuni')
# Agregar datos de municipios al DataFrame de estados 'join'
join = pd.concat([join, top_municipios])


# Calcular la cantidad total de residuos generados por cada zona
total_norte = join[join['Zona'] == 'Norte']['Total'].sum()
total_centro = join[join['Zona'] == 'Centro']['Total'].sum()
total_sur_sureste = join[join['Zona'] == 'Sur']['Total'].sum()

# Crear leyenda de colores para las zonas
legend_labels = [f'Total Norte: {total_norte}', f'Total Centro: {total_centro}', f'Total Sur Sureste: {total_sur_sureste}']

x = [point.x for point in join.geometry.centroid]
y = [point.y for point in join.geometry.centroid]
# Plot the map
fig, ax = plt.subplots(figsize=(20, 8))

# Plot the map with different colors for each zone
join.plot(column='Zona', cmap='Reds', ax=ax)

heatmap = sns.kdeplot(
    # data=join.geometry,
    x=x,
    y=y, 
    cmap='YlOrBr',  # Cambiar el mapa de color a amarillo-anaranjado
    shade=True, 
    bw_adjust=0.4,  # Ajuste del ancho de banda
    alpha=0.6,      # Transparencia
    ax=ax
)

# for idx, row in top_municipios.iterrows():
#     ax.annotate(text=row['NOMGEO'], xy=(row.geometry.centroid.x, row.geometry.centroid.y), color='blue', fontsize=8, ha='center')

legend_handles = []
for idx, row in top_municipios.iterrows():
    legend_handles.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=5, label=row['NOMGEO']))

# Plot top 10 municipalities as blue points
top_municipios.plot(color='blue', ax=ax)

plt.title('Mapa de calor de Generación de Residuos en México')

# Calcular la densidad de kernel gaussiano para la generación de residuos
kde = gaussian_kde([x, y], bw_method=0.4)

# Definir los límites del mapa
x_min, x_max = min(x), max(x)
y_min, y_max = min(y), max(y)

# Generar una cuadrícula de puntos
x_grid = np.linspace(x_min, x_max, 100)
y_grid = np.linspace(y_min, y_max, 100)
x_mesh, y_mesh = np.meshgrid(x_grid, y_grid)
grid_coords = np.vstack([x_mesh.ravel(), y_mesh.ravel()])

# Calcular la densidad de kernel gaussiano en la cuadrícula
density_values = kde(grid_coords)

# Reorganizar los valores de densidad en una matriz 2D
density_matrix = density_values.reshape(x_mesh.shape)

# Plotear el mapa de calor
heatmap = ax.imshow(density_matrix, cmap='YlOrBr', extent=[x_min, x_max, y_min, y_max], aspect='auto', origin='lower')

# Plotear los municipios sobre el mapa de calor
top_municipios.plot(ax=ax, color='blue')

# Añadir barra de color para el mapa de calor
cbar = plt.colorbar(heatmap, ax=ax, label='Densidad', cmap='YlOrBr')
# Añadir leyenda de municipios
legend_elements = []
for idx, row in top_municipios.iterrows():
    legend_elements.append(plt.Line2D([0], [0], marker='o', color='blue', markersize=5, label=row['NOMGEO']))
plt.legend(handles=legend_elements, loc='upper left', title='Municipios')
ax.set_axis_off()
# Añadir título
plt.title('Mapa de calor de Generación de Residuos en México')

plt.show()
```
